```python
df = pd.read_csv('../Lung_Paper_Images/Test_Images/CAPA4_Validation.csv', comment='#')
df.head()
```

```python
inds = df['Tissue'] == 'Lung'
df_lung = df.loc[inds, :]
#df_lung = df_lung.drop(columns = ['Count 2'])
df_lung.reset_index()
```

```python
import bokeh
from bokeh.models.widgets import Panel, Tabs
from bokeh.io import output_file, show
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.sampledata.commits import data
from bokeh.transform import jitter
```

```python
#Initialize an output file to save an html of the generated plot
#output_file("../Lung_Paper_Images/Quantification/Lung/lung_transduction_per_sq_mm.html")

# Store the data for plotting
source = ColumnDataSource(df_lung)

# Intitialize a figure
p = figure(plot_width=800, plot_height=800, x_axis_type = 'log', y_axis_type = 'log', x_axis_label = 'Hand Count of Cells', y_axis_label = 'Automated Count', title = 'Lung Accuracy')

# Plot the data as a scatter plot

p.circle(x = 'Actual Number of Cells', y='Count 5', source = source, alpha=1, color = '#e6ab02', size = 5, legend = 'Intensity Threshold = 100, Shape Threshold = 0.62')
p.circle(x = 'Actual Number of Cells', y='Count 1', source = source, alpha=1, color = '#1b9e77', size = 5, legend = 'Intensity Threshold = 1000, Shape Threshold = 0.62')
p.circle(x = 'Actual Number of Cells', y='Count 2', source = source, alpha=1, color = '#d95f02', size = 5, legend = 'Intensity Threshold = 10000, Shape Threshold = 0.62')
p.circle(x = 'Actual Number of Cells', y='Count 3', source = source, alpha=1, color = '#7570b3', size = 5, legend = 'Intensity Threshold = 1000, Shape Threshold = 0.5')
p.circle(x = 'Actual Number of Cells', y='Count 4', source = source, alpha=1, color = '#e7298a', size = 5, legend = 'Intensity Threshold = 1000, Shape Threshold = 0.65')



p.line([1, 500], [1, 500], line_width=1, legend = 'Perfect Correlation')


# Simply modify plot characteristics 
p.plot_height = 800
p.plot_width = 800

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

p.xaxis.minor_tick_line_color = None
p.yaxis.minor_tick_line_color = None

p.xaxis.major_label_text_font_size = '11pt'
p.yaxis.major_label_text_font_size = '11pt'

p.xaxis.axis_label_text_font_size = '11pt'
p.yaxis.axis_label_text_font_size = '11pt'

p.background_fill_color = None 
p.border_fill_color = None

p.legend.location = 'top_left'

p.output_backend = "svg"

show(p)
```
